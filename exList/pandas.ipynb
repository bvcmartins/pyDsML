** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
(sal.groupby("JobTitle").size().sort_values()==1).sum()
```

```python
# filter first using loc
# then groupby
# then reduce
(sal.loc[sal["Year"]==2013].groupby("JobTitle").size().sort_values()==1).sum()
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
import re
sal[sal["JobTitle"].str.contains('Chief')==True]
```

```python
sal[sal["JobTitle"].str.contains('^Chief*',case=True,regex=True)==True]
```

```python
sal[sal["JobTitle"].str.contains('Chief|chief|CHIEF',case=True,regex=True)==True]["Id"].count()
```

```python
sal[sal["JobTitle"].str.match('Chief',case=True)==True]["Id"].count()
```
