```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
# functional method to create plot
plt.plot(x,y,'r-')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b.')
```

```python
# Objected oriented
fig = plt.figure() # define a blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # manually choose the axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # insert plot
# numbers are fractions of the main canvas
axes1.plot(x,y)
axes1.set_title("Plot 1")
axes2.plot(y,x)
axes2.set_title("Plot 2")
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) # for mulitple plots
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) # for mulitple plots
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(y,x)
axes[1].set_title('Plot 2')
plt.tight_layout()
```

### Figure size and dpi

```python
fig = plt.figure(figsize=(5,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(5,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('myPic.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label="Plot 1")
ax.set_title("Fig 1")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.plot(x,2*y, label="Plot 2")
#ax.legend(loc=(0.1,0.1))
ax.legend(loc='best')
plt.tight_layout()
```
